### Now applying the test case to all models

```python
df_train = pd.read_pickle("../data/processed/clean_df.pkl")
authors = df_train['authorName'].unique()
```

```python
scores = np.zeros(len(authors))
```

```python
len(authors)
```

```python
for i in range(len(authors)):
        author = authors[i]

        with open(f"../models/{author}.pkl", 'rb') as f:
                models = pickle.load(file = f)

        NLP_model = models[0]

        modelcount = len(NLP_model.get_feature_names_out())
        testcount = NLP_model.transform([testcase]).count_nonzero()
        scores[i] = testcount / modelcount
        print(i)
```

```python
predictions = pd.DataFrame(zip(authors,scores),columns = ["author","score"])
```

```python
predictions.head(10)
```

### now apply place model to all cases

```python
df_test = df_train[df_train['authorName']=='Chuhan Wu']
df_place_in = df_test[["year","venues_le"]]
df_place_year = df_place_in["year"].tolist()
df_place_venue = df_place_in["venues_le"].tolist()
testcase_place = [df_place_year[0], df_place_venue[0]]
testcase_place
```

```python
import warnings
warnings.filterwarnings('ignore') 
```

```python
place_scores = np.zeros(len(authors))
```

```python
for i in range(len(authors)):
        author = authors[i]

        with open(f"../models/{author}.pkl", 'rb') as f:
                models = pickle.load(file = f)

        place_model = models[1]
        place_scores[i] = place_model.predict_proba([testcase_place])[0][1]

        print(i,scores[i])
```

```python
place_scores_series = pd.Series(place_scores)
```

```python
place_scores_series.describe()
```

### Combining results

```python
total_predictions = predictions 
```

```python
total_predictions["score_place"] = place_scores_series
```

```python
total_predictions["combined_score"] = total_predictions["score_place"] * total_predictions["score"]
```

```python
sorted = total_predictions.sort_values("combined_score",ascending=False)
sorted.head(20)
```
